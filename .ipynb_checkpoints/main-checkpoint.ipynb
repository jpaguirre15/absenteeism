```python
# Get dummies for 'Reason for absence'
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
# Create a 'check' column 
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns
```

```python
reason_columns['check'].unique()
```

```python
reason_columns = reason_columns.drop(['check'], axis=1)
reason_columns
```

```python
# Remove 0 index to avoid multi-collinearity
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_columns
```

```python
# Drop 'Reason for Absence' from df
df = df.drop(['Reason for Absence'], axis=1)
```

```python
# Confirm that 'Reason for Absence' is deleted
df.info()
```

```python
# Group reasonings
reason_group_1 = reason_columns.loc[:, 1:14].max(axis = 1)
reason_group_2 = reason_columns.loc[:, 15:17].max(axis = 1)
reason_group_3 = reason_columns.loc[:, 18:21].max(axis = 1)
reason_group_4 = reason_columns.loc[:, 22:28].max(axis = 1)
```

```python
# Concatenate Reason groups
df = pd.concat([df, reason_group_1, reason_group_2, reason_group_3, reason_group_4], axis=1)
df
```

```python
df.columns.values
```

```python
column_names = ['ID', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
```

```python
df
```

```python
# Reorder Columns
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[column_names_reordered]
```

```python
# Check reordered columns
df.head()
```

```python
# Create a Checkpoint
df_reason_mod = df.copy()
```

```python
df_reason_mod
```

```python
# Convert 'Date' object Series to datetime 
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'])
```

```python
type(df_reason_mod['Date'][0])
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format='%d/%m/%Y')
```

```python
type(df_reason_mod['Date'][0])
```

```python
# Extract Month value from date

# Preview Timestamp at index 0
df_reason_mod['Date'][0]
```

```python
# Preview Timestamp month at index 0 
df_reason_mod['Date'][0].month
```

```python
list_months = []

for i in range(len(df_reason_mod['Date'])):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
df_reason_mod['Month Value'] = list_months
```

```python
df_reason_mod.head(10)
```

```python
# Extract the Day of the Weeek

df_reason_mod['Date'][699].weekday()
```

```python
# This will show that this date is a Thursday (0 starts at Monday)
df_reason_mod['Date'][699]
```

```python
# Create a function to extract Day of the Week from Timestamp
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
# Apply the day of the week
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
# Exercise 6 
# Do the same thing with the ‘Date’ column – drop it from the df_reason_mod 
# DataFrame.

# Remove 'Date' column 
df_reason_mod = df_reason_mod.drop(['Date'], axis=1)
```

```python
df_reason_mod.columns.values
```

```python
# Move 'Month Value' and 'Day of the Week' between 'Reasons ..' to 'Trans. Exp.' 
df_reason_mod_columns = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Month Value',
       'Day of the Week',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df_reason_mod = df_reason_mod[df_reason_mod_columns]
```

```python
df_reason_mod.head()
```

```python
df_reason_mod.info()
```

```python
# Checkpoint
df_reason_date_mod = df_reason_mod.copy() 
```

```python
# Education column Analysis
df_reason_date_mod['Education'].unique()
```

```python
df_reason_date_mod['Education'].value_counts()
```

```python
df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
df_reason_date_mod['Education'].unique()
```
